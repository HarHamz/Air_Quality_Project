### Assessing Data

#### Counting the Faulty Data

Let's count empty data, duplicated data, and outliers. We will use `describe()`, `isna().sum()`, `duplicated().sum()`, and a `count_outliers` function.

```python
def count_outliers(df, columns):
    Q1 = (df[columns].quantile(0.25))
    Q3 = (df[columns].quantile(0.75))
    IQR = Q3 - Q1
 
    maximum = Q3 + (1.5*IQR)
    minimum = Q1 - (1.5*IQR)

    return (df[columns] < minimum).sum() + (df[columns] > maximum).sum()


columns = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "WSPM"]
```

## Exploratory Data Analysis (EDA)

### Exploring the Minimum values, Maximum values, Average values, Q1, Median, and Q3 of the temperatures in all the datasets

## Visualization & Explanatory Analysis

### Question 1: At which stations were the highest and lowest temperatures recorded?

```python
min_temp = df_all.loc[df_all.groupby("station")["TEMP"].idxmin(), ["station", "TEMP"]].rename(columns={"TEMP": "Min Temp"})
max_temp = df_all.loc[df_all.groupby("station")["TEMP"].idxmax(), ["station", "TEMP"]].rename(columns={"TEMP": "Max Temp"})

df_extremes = min_temp.merge(max_temp, on="station")
df_extremes
```

```python
stations = df_extremes["station"]

x = np.arange(len(stations))
width = 0.4

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(x - width/2, df_extremes["Min Temp"], width, label="Min Temp", color="blue", alpha=0.7)
ax.bar(x + width/2, df_extremes["Max Temp"], width, label="Max Temp", color="red", alpha=0.7)

ax.set_xlabel("Station", fontsize=12)
ax.set_ylabel("Temperature (C)", fontsize=12)
ax.set_title("Maximum and Minimum Temperatures by Station", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(stations, rotation=45, ha="right")
ax.legend()
ax.grid(axis="y", linestyle="--", alpha=0.5)

bars_min = ax.bar(x - width/2, df_extremes["Min Temp"], width, label="Min Temp", color="blue", alpha=0.7)
bars_max = ax.bar(x + width/2, df_extremes["Max Temp"], width, label="Max Temp", color="red", alpha=0.7)

for bar in bars_min:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f"{height:.1f}", ha="center", va="top", fontsize=10, color="blue")
for bar in bars_max:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f"{height:.1f}", ha="center", va="bottom", fontsize=10, color="red")

plt.tight_layout()
plt.show()
```

**Insight:**
The highest temperature recorded was in Gucheng at 41.6 C, and the lowest was in Huairou at -19.9 C. Additionally, the maximum and minimum temperatures across all weather stations are relatively similar, as the stations are located in close proximity to one another.

### Question 2: How does the temperature change at each station every month in 2016?

To answer this question, we will use line charts for each of the stations' mean recorded temperature for every month in 2016.

```python
df_average_temp_2016 = df_all.query("year == 2016").groupby(["station", "month"], as_index=False)["TEMP"].mean().rename(columns={"TEMP": "Average Temp"})

df_average_temp_2016
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_average_temp_2016, x="month", y="Average Temp", hue="station")

plt.title("Average Temperature by Station at 2016")
plt.xlabel("Month")
plt.ylabel("Average Temperature (C)")
plt.legend(loc="lower center")
plt.grid(True)

# setting xticks to show months
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
plt.xticks(ticks=range(1, 13), labels=months, rotation=45)

plt.show()
```

**Insight:** In 2016, every stations experience their lowest temperatures in January, and highest in July - August.

### Question 3: Is there a pattern for the concentration of SO2, NO2, CO, and O3 throughout the year in 2016?

```python
df_average_part_2016 = df_all.query("year == 2016").groupby(["station", "month"], as_index=False)[["SO2", "NO2", "CO", "O3"]].mean()

df_average_part_2016
```

```python
pollutants = ["SO2", "NO2", "CO", "O3"]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Monthly Average Pollutant Concentrations by Station (2016)", fontsize=16)

axes = axes.flatten()

for i, pollutant in enumerate(pollutants):
    sns.lineplot(
        data=df_average_part_2016,
        x="month",
        y=pollutant,
        hue="station",
        ax=axes[i],
        marker="o"
    )
    axes[i].set_title(f"{pollutant} Concentration")
    axes[i].set_xlabel("Month")
    axes[i].set_ylabel(f"{pollutant} Concentration (g/m)")
    axes[i].set_xticks(ticks=range(1, 13), labels=months, rotation=45)
    axes[i].grid(True)
    axes[i].legend(loc="upper right")

plt.tight_layout()
plt.show()
```

**Insight:** In 2016, every stations experience the same patterns for the amount of pollutants.
- SO2 concentration are highest in January and December, with a spike in March, while June - October experiences the lowest amounts.
- NO2 concentration are highest in December, with a significant spike in March.
- CO concentration are highest in January and December, with a spike in March, while April - September experiences the lowest amounts.
- O3 concentration are nearly the opposite of the other pollutants, showing a peak in June, while January and October - December experience the least amounts.

### Question 4: Is there a correlation between the concentration of SO2, NO2, CO, O3, and temperatures recorded in 2016 at Shunyi?

We will use the previous `df_average_temp_year` and `df_average_part_year` dataset to see whether or not there is a correlation between the concentration of SO2, NO2,  CO, and O3 levels with temperatures using the correlation heatmap.

```python
df_combined_2016 = pd.merge(
    df_average_part_2016,
    df_average_temp_2016,
    on=["station", "month"]
)

plt.figure(figsize=(8, 6))
sns.heatmap(df_combined_2016[["SO2", "NO2", "CO", "O3", "Average Temp"]].corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)

plt.title("Pollutants to Temperature Correlation Heatmap in Shunyi")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

**Insight:**
- There is a relatively strong positive correlation between the levels of SO, NO, and CO
- There is a relatively strong negative correlation between O and the other pollutants (SO, NO, and CO).
- There is an exceptionally strong positive correlation between O and average temperatures.
- There is an exceptionally strong negative correlation between SO and average temperature.
- There is a relatively strong negative correlation between NO and CO and average temperature.

## Additional Analysis

#### Explaining the correlations of the pollutants and temperature

According to Ngarambe, et. al. in their research "[Exploring the relationship between particulate matter, CO, SO2, NO2, O3 and urban heat island in Seoul, Korea](https://www.sciencedirect.com/science/article/abs/pii/S0304389420316010)", it was found that carbon monoxide (CO), nitrogen dioxide (NO), sulfur dioxide (SO), and particulate matter (PM) exhibit positive correlations with urban heat island intensity (UHII). In contrast, ozone (O) showed a negative correlation with UHII.
This confirms the correlation between the pollutants and temperature that we found, where SO, NO, and CO have positive correlation with temperature, while O has a negative correlation with temperature.
